```python
def beta_diversity_bars(dm, meta, category, order=None, ref_group=None,
    p_table=None, num_iter=999, p_crit=0.01,
    ax=None, interval=0.1, width=0.1, bar_props={}):

    # Removes any undefined groups
    map_ = meta.groupby(meta[category].apply(check_strs)).get_group(True)
    dm = dm.filter(map_.index)

    # Orders the objects
    if order is None:
        order = map_.groupby(category).groups.keys()
    if ref_group is None:
        ref_group = order[0]

    # Determines hte position of the reference group
    if ref_group == 'all_within':
        ref_loc = 0
    else:
        ref_loc = order.index(ref_group)

    # Checks for an overall signifigant differences
    perma_res = permanova(dm, map_, category, num_iter)
    all_p = perma_res['p-value']
    if all_p > p_crit:
        raise ValueError('There is not a signfigant difference at p < %s.'
                         '\np = %1.2f' % (p_crit, all_p))
    if 'p_value' not in bar_props:
        bar_props['p_value'] = all_p

    # Gets the distance vectors
    w_groups, w_dist, b_groups, b_dist = \
        get_distance_vectors(dm, meta, category, order)

    # Gets the distances
    if ref_group == 'all_within':
        dist_bar = [np.hstack(w_dist).mean()]
        dist_std = [np.hstack(w_dist).std()]
    else:
        dist_bar = [w_dist[w_groups.index(ref_group)].mean()]
        dist_std = [w_dist[w_groups.index(ref_group)].std()]

    if p_table is None:
        p_values = p_table
    else:
        p_values = np.array([])
        sub_p = pd.concat([p_table.loc[p_table['Group 1'] == '%s vs. %s'
                                       % (ref_group, ref_group)],
                           p_table.loc[p_table['Group 2'] == '%s vs. %s'
                                       % (ref_group, ref_group)]])

    for idx, group in enumerate(order[:-1]):
        if ref_group == 'all_within':
            dist_bar.extend([np.zeros(1)])
            dist_bar.extend([b_dist[b_groups.index[{group, g2}]].mean() for g2 in order[idx+1:]])
            dist_std.extend([np.zeros(1)])
            dist_std.extend([b_dist[b_groups.index[{group, g2}]].std() for g2 in order[idx+1:]])
        elif group == ref_group:
            continue
        else:
            dist_bar.append(b_dist[b_groups.index({ref_group, group})].mean())
            dist_std.append(b_dist[b_groups.index({ref_group, group})].std())
    return dist_bar, dist_std
```

We'll use two tests: the kruskal-wallis test, which is a rank-sum test and compare abundance, and the g-test, which is functionally simillar to the chi-square test and examines the presence/absence of taxa.
