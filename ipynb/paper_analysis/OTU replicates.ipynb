We can start by writing a function which will repeatedly calculate rarefaction over the different depths, and the average 

```python
def calculate_rare_samples(table, depth, samples, num_iter=5):
    """Determines the number of otus from a set of rarefactions"""
    depth_watch = np.zeros((num_iter, len(samples)))

    # Preforms the rarefactions
    for id1 in xrange(num_iter):
        for id2, samp in enumerate(samples[::-1]):
            sub1 = table.copy()
            sub1 = sub1.subsample(depth)
            if sub1.shape[1] < samp:
                depth_watch[id1, id2] = np.nan
                continue
            sub1 = sub1.subsample(samp, by_id=True)
            depth_watch[id1, len(samples) - (id2 + 1)] = sub1.shape[0]

    return depth_watch
```

We can now get the rarefation matrix, using our function. We'll start by prealocating these results.

```python
# Prealocates a holding array
awt_count_mean = np.zeros((num_rarefactions, num_samples))
awt_count_stdv = np.zeros((num_rarefactions, num_samples))
ast_count_mean = np.zeros((num_rarefactions, num_samples))
ast_count_stdv = np.zeros((num_rarefactions, num_samples))
ass_count_mean = np.zeros((num_rarefactions, num_samples))
ass_count_stdv = np.zeros((num_rarefactions, num_samples))
hmp_count_mean = np.zeros((num_rarefactions, num_samples))
hmp_count_stdv = np.zeros((num_rarefactions, num_samples))

for id1, depth in enumerate(rare_depths):
    print depth
    # Gets the OTU counts for various sample iterations
    awt_depth = calculate_rare_samples(awt_otu, depth, samps_drawn, num_iterations)
    awt_count_mean[id1, :] = awt_depth.mean(0)
    awt_count_stdv[id1, :] = awt_depth.std(0)
    ast_depth = calculate_rare_samples(ast_otu, depth, samps_drawn, num_iterations)
    ast_count_mean[id1, :] = ast_depth.mean(0)
    ast_count_stdv[id1, :] = ast_depth.std(0)
    ass_depth = calculate_rare_samples(ass_otu, depth, samps_drawn, num_iterations)
    ass_count_mean[id1, :] = ass_depth.mean(0)
    ass_count_stdv[id1, :] = ass_depth.std(0)
    hmp_depth = calculate_rare_samples(fhmp_otu, depth, samps_drawn, num_iterations)
    hmp_count_mean[id1, :] = hmp_depth.mean(0)
    hmp_count_stdv[id1, :] = hmp_depth.std(0)
```
