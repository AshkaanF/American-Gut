We can also try controlling for a set of variables. Let's control for 
<code>IBD</code>, <code>BMI_CAT</code>, <code>TYPES_OF_PLANTS</code>, <code>DIABETES</code>, <code>AGE_CAT</code> and <code>SLEEP_DURATION</code>.
We will also clean up the sleep duration, combining people who have slept less than 5 hours and people who have slept 5-6 becuase otherwise, the sample group will be insuffecient. We will test this for a seasonal effect, becuase there are not enough observations in each group otherwise.

```python
# Maps sleep duration so that less than 6 hours are combined
aws_map.loc[aws_map.SLEEP_DURATION == 'Less than 5 hours', 'SLEEP_DURATION'] = 'Less than 6 hours'
aws_map.loc[aws_map.SLEEP_DURATION == '5-6 hours', 'SLEEP_DURATION'] = 'Less than 6 hours'

# Gets the paired samples
control_categories = ['IBD', 'BMI_CAT', 'TYPES_OF_PLANTS', 'DIABETES', 
                      'AGE_CAT', 'SLEEP_DURATION']

aws_map.groupby(['IBD', 'BMI_CAT', 'AGE_CAT', 'TYPES_OF_PLANTS', 
                 'DIABETES', 'SLEEP_DURATION'])

season_paired_ids = get_paired_subsamples(aws_map, season, control_categories, order=s_order)
# Checks the number of ids
print 'There are %i paired samples.' % season_paired_ids[0].shape[0]

# Rearranges the paired samples back into a list
season_ids = season_paired_ids[0]
for ids in season_paired_ids[1:]:
    season_ids = hstack((season_ids, ids))

# Filters the table down to the samples
season_paired_table = aws_map.loc[season_ids]
```

```python
# Plots the alpha diversity data
paired_season_a_ax, paired_season_a_feats = \
    pretty_pandas_boxplot(season_paired_table, 
                          season, 
                          s_order, 
                          a_div_metric,
                          boxplot_props= {'ylabel': a_ylabel,
                                           'title': "Season, Controlled.",
                                           'hide_xticks': True,
                                           'ylims': [10, 55],
                                           'p_y': 51.5})
paired_season_a_ax.set_position((0.125, 0.125, 0.45, 0.7))
# Preforms a post-hoc test
paired_season_a_post = post_hoc_pandas(season_paired_table.groupby(season),
                                 order=s_order,
                                 cat=a_div_metric)
paired_season_a_post
```

We can then import the compared beta diveristy results, and plot the distance for each of the tables, starting with unweighted UniFrac for the unfiltered table.

```python
# Creates a boxplot of the beta diversity
ass_season_ub_ax, ass_season_ub_feats = \
    compare_beta_diversity(ass_ubd, ass_map, season, num_iter=999,
                           show_allbetween=False, show_within=False,
                           violin_props={'ylabel': 'Unweighted UniFrac',
                                          'show_n': False,
                                          'title': 'Collection Season, healthy subset',
                                          'hide_xticks': True,
                                          'ylims': [0.3, 1],
                                          'xfont_angle': 45,
                                          'xfont_align': 'right'})

# Calculates the mean distance
sub_ids = (ass_season_ub_post['Group 1'].apply(keep_f) | ass_season_ub_post['Group 2'].apply(keep_f)) & \
          (ass_season_ub_post['Group 1'].apply(discard_f) & ass_season_ub_post['Group 2'].apply(discard_f))

ass_season_ub_post.loc[sub_ids, ['Group 1', 'Group 2', 'Parametric p-value (Bonferroni-corrected)']]
```

```python
# Creates a boxplot of the beta diversity
ass_season_wb_ax, ass_season_wb_feats = \
    compare_beta_diversity(ass_wbd, ass_map, season, num_iter=999,
                           show_allbetween=False, show_within=False,
                           violin_props={'ylabel': 'Weighted UniFrac',
                                          'show_n': False,
                                          'title': 'Collection Season, healthy subset',
                                          'hide_xticks': True,
                                          'xfont_angle': 45,
                                          'xfont_align': 'right'})

# Calculates the mean distance
sub_ids = (ass_season_wb_post['Group 1'].apply(keep_f) | ass_season_wb_post['Group 2'].apply(keep_f)) & \
          (ass_season_wb_post['Group 1'].apply(discard_f) & ass_season_wb_post['Group 2'].apply(discard_f))
ass_season_wb_post.loc[sub_ids, ['Group 1', 'Group 2', 'Parametric p-value (Bonferroni-corrected)']]
```
